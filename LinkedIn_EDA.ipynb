Import the Necessary Libraries

```python
import pandas as pd
import plotly.express as px
```

Data Preparation

```python
# Read the CSV Data
df = pd.read_csv('data/final_data.csv')

# List of skills based on your dataset columns
skill_cols = [
    'PYTHON', 'C++', 'JAVA', 'HADOOP', 'SCALA', 'FLASK', 'PANDAS', 'SPARK', 
    'NUMPY', 'PHP', 'SQL', 'MYSQL', 'CSS', 'MONGODB', 'NLTK', 'TENSORFLOW', 
    'LINUX', 'RUBY', 'JAVASCRIPT', 'DJANGO', 'REACT', 'REACTJS', 'AI', 'UI', 
    'TABLEAU', 'NODEJS', 'EXCEL', 'POWER BI', 'SELENIUM', 'HTML', 'ML'
]

# Calculate the frequency of each skill
demand_skills = df[skill_cols].sum().reset_index(0)
demand_skills.columns = ['Skill', 'Count']
demand_skills = demand_skills.sort_values('Count', ascending=False)
```

Data Visualization


```python
fig_bar = px.bar(
    demand_skills,
    x='Count',
    y='Skill',
    title='Top Demanded Skills in tech Job',
    orientation='h',
    color='Count',
    color_continuous_scale='Viridis',
    height=800
)

fig_bar.show()
```
